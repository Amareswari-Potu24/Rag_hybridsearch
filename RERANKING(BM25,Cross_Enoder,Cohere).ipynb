```python
documents=["Albert Einstein was a theoretical physicist who developed the theory of relativity.",
    "Isaac Newton formulated the laws of motion and universal gravitation.",
    "Marie Curie conducted pioneering research on radioactivity and won two Nobel Prizes.",
    "Nikola Tesla was known for his contributions to the design of the modern AC electricity supply system.",
    "Stephen Hawking was a theoretical physicist who worked on black holes and cosmology."
    ]
```

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
```

```python
#loading pretrained sentence transformer model
model_name="sentence-transformers/paraphrase-xlm-r-multilingual-v1"
```

```python
model=SentenceTransformer(model_name)
```

```python
documents
```

```python
len(documents)
```

```python
document_embeddings=model.encode(documents)
```

```python
len(document_embeddings)
```

```python
len(document_embeddings[0])
```

```python
for i, embedding in enumerate(document_embeddings):
  print(f"Document {i+1} Embedding: {embedding}")
```

```python
query="Who developed the theory of relativity?"
```

```python
quer_embedding=model.encode(query)
```

```python
quer_embedding
```

```python
len(quer_embedding)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities=cosine_similarity(np.array([quer_embedding]),document_embeddings)
```

```python
similarities
```

```python
most_similar_index=np.argmax(similarities)
```

```python
most_similar_index
```

```python
most_similar_document=documents[most_similar_index]
```

```python
most_similar_document
```

```python
query
```

```python
similarity_score=similarities[0][most_similar_index]
```

```python
similarity_score
```

```python
sorted_indices=np.argsort(similarities[0])[::-1]
```

```python
sorted_indices
```

```python
ranked_documents=[(documents[i],similarities[0][i]) for i in sorted_indices]
```

```python
ranked_documents
```

```python
print("Ranked Documents:")
for rank, (document, similarity) in enumerate(ranked_documents, start=1):
    print(f"Rank {rank}: Document - '{document}', Similarity Score - {similarity}")
```

```python
print("Top 4 Documents:")
for rank, (document, similarity) in enumerate(ranked_documents[:4], start=1):
    print(f"Rank {rank}: Document - '{document}', Similarity Score - {similarity}")
```

```python
from rank_bm25 import BM25Okapi
```

```python
top_4_documents=[doc[0] for doc in ranked_documents[:4]]
```

```python
top_4_documents
```

```python
tokenized_top_4_document=[doc.split() for doc in top_4_documents]
```

```python
tokenized_top_4_document
```

```python
tokenized_query=query.split()
```

```python
tokenized_query
```

```python
bm25=BM25Okapi(tokenized_top_4_document)
```

```python
bm25
```

```python
bm25_scores=bm25.get_scores(tokenized_query)
```

```python
bm25_scores
```

```python
sorted_indices2=np.argsort(bm25_scores)[::-1]
```

```python
sorted_indices2
```

```python
top_4_documents
```

```python
reranked_documents=[(top_4_documents[i],bm25_scores[i]) for i in sorted_indices2]
```

```python
reranked_documents
```

```python
from sentence_transformers import CrossEncoder
```

```python
cross_encoder=CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")
```

```python
top_4_documents
```

```python
pairs=[]
for doc in top_4_documents:
  pairs.append([query,doc])
```

```python
pairs
```

```python
scores=cross_encoder.predict(pairs)
scores
```

```python
scored_docs=zip(scores,top_4_documents)
```

```python
scored_docs
```

```python
reranked_document_cross_encoder=sorted(scored_docs,reverse=True)
```
